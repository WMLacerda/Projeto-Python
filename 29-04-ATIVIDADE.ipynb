Operador or (OU) → pelo menos uma verdadeira

2. Verifique se o usuário digitou "sim" ou "s".
3. Verifique se o aluno passou (nota >= 7 ou fez recuperação).
4. Verifique se o número é par ou múltiplo de 5.
5. Verifique se a pessoa é idosa (>=60) ou menor de idade (<18).


Operador not (NÃO) → inverte o valor lógico
1. Verifique se o usuário não digitou "admin".
2. Verifique se um número não é positivo.
3. Verifique se o aluno não foi aprovado (nota < 7).
4. Verifique se um número não é par.
5. Verifique se a senha está incorreta.


```python
#Operador and (E) → ambas verdadeiras
#2. Verifique se o usuário tem idade maior que 18 e possui carteira de motorista.

idade = int(input("Digite sua idade: "))
cnh = input("Vc possui carteira de motorista? ").lower()

if idade >= 18 and cnh == "sim":
  print("Usuário maior que 18 anos e possui CNH")
else:
  print("Não é maior que 18 anos ou não possui CNH")
```

```python
#3. Verifique se a senha está correta e o usuário é "admin".
usuario = input("Digite seu usuário: ")
senha = input("Digite sua senha: ")
if usuario == "admin" and senha == "1234":
  print("Acesso permitido")
else:
  print("Acesso negado! Senha ou usuário incorreto")
```

```python
# 4. Verifique se um aluno passou (nota >= 7 e frequência >= 75%).

nota = float(input("Digite sua nota: "))
frequencia = int(input("Digite sua frequência em porcentagem: "))

if nota >= 7 and frequencia >= 75:
  print("Vc passou!")
else:
  print("Vc reprovou")
```

```python
#5. Verifique se um número é positivo e par.

num = int(input("Digite um número interio: "))
if num >= 1 and num % 2 == 0:
  print("O número digitado é positivo e par")
else:
  print("Ou o número não é positivo ou não é par")
```

```python
#Verifique se um número é menor que 0 ou maior que 100.

num = float(input("Digite um número: "))
if num < 0 or num > 100:
  print("Némro menor que 0 ou maior que 100")
else:
  print("Número entre 0 e 100")
```

```python
#Verifique se o usuário digitou "sim" ou "s".

usuario = input("Digite 'sim' ou 's': ").lower()
if usuario == "sim" or usuario == "s":
  print("Usuário digitou 'sim' ou 's'")
else:
  print("Usuário não digitou 'sim' nem 's' ")
```

```python
#Verifique se o aluno passou (nota >= 7 ou fez recuperação).

nota = float(input("Digite sua nota: "))
recuperacao = input("fez recuperação 's/n'? ").lower()

if nota >= 7 or recuperacao == 's':
  print("aluno passou!")
else:
  print("Aluno reprovou!")
```

```python
#Verifique se o número é par ou múltiplo de 5.

num = int(input("Digite um número inteiro: "))


if num  % 2 == 0 or num % 5 == 0:
  print("O número digitado é par ou multiplo de 5")
else:
  print("O número digitado não é par ou não é multiplo de 5")
```

```python
#Verifique se a pessoa é idosa (>=60) ou menor de idade (<18).
idade = int(input("Digite sua idade: "))
if idade >= 60 or idade < 18:
  print("Vc é idoso ou menor de idade.")
else:
  print("Vc é um jovem que está entre 18 anos e 59 anos ")
```

```python
#Operador not (NÃO) → inverte o valor lógico

#Verifique se o usuário não digitou "admin".

usuario = input("Digite seu login: ")

if not usuario ==  "admin":
  print("Não digitou 'admin'.")
else:
  print("Digitou 'admin'.")
```

```python
#Verifique se um número não é positivo.

num = int(input("Digite um número inteiro: "))
if not num >= 1:
  print("O número não é positivo")
else:
  print("O número é positivo ou é 0")
```

```python
#Verifique se o aluno não foi aprovado (nota < 7).

nota = float(input("Digite sua nota: "))

if not nota < 7:
  print("Vc foi aprovado sua nota é Maior ou igual a 7")
else:
  print("vc foi reprovado Sua nota é menor que 7")
```

```python
#Verifique se um número não é par.

num = int(input("Digite um número inteiro"))
if not num % 2 == 0:
  print("O número não é par")
else:
  print("O número é par")
```

```python
#Verifique se a senha está incorreta.

senha = input("Digite sua senha: ")

if not senha == "wemerson@2026":
  print("senha correto")
else:
  print("Senha incorreta")
```
